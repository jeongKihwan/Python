```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cust_df = pd.read_csv('./files/train_santander.csv')
cust_df.head()
```

```python
cust_df.info()
```

```python
cust_df.TARGET.value_counts()
```

```python
cust_df.describe()
```

```python
(cust_df['var3'] == -999999.000000).sum()
```

```python
cust_df['var3'] = (cust_df['var3'].replace(-999999.000000, 2)).min()
cust_df['var3'] = (cust_df['var3'].replace(-999999, 2)).min()
```

```python
cust_df.describe()
```

```python
cust_df.drop('ID', axis=1, inplace=True)
```

```python
# cust_df.describe()
```

```python
y = cust_df['TARGET']
x = cust_df.drop('TARGET', axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=111)
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import roc_auc_score
```

```python
xgb_cls = XGBClassifier(n_estimators=500, random_state=155)
```

```python
xgb_cls.fit(x_train, y_train, eval_metric= 'auc', eval_set=[(x_train, y_train), (x_test, y_test)])
```

```python
roc_auc_score(y_test, xgb_cls.predict_proba(x_test)[:,1], average='macro')
```

```python
# 성과지표가 더 이상 개션이 되지 않으면 중간에 학습을 중단시킬 수 있다.
# early stopping -  주어진 횟수만큼 계속 성과개선이 되지 않으면 학습을 중단한다...
```

```python
xgb_cls.fit(x_train, y_train, eval_metric= 'auc', early_stopping_rounds = 50, \
            eval_set=[(x_train, y_train), (x_test, y_test)])
```

```python
roc_auc_score(y_test, xgb_cls.predict_proba(x_test)[:,1], average='macro')
```

```python
x_train.columns
```

```python
# parameter최적화를 통한 성과개선..
```

```python
xgb_cls = XGBClassifier(n_estimators=100, random_state=155)
```

```python
params = {'max_depth':[5,7,9], 'min_child_weight':[1,3], 'colsample_bytree':[0.5,  0.75]}
```

```python
from sklearn.model_selection import GridSearchCV

grid = GridSearchCV(xgb_cls, param_grid = params)
```

```python
grid.fit(x_train, y_train, eval_metric='auc', early_stopping_rounds = 30\
         , eval_set=[(x_train, y_train), (x_test, y_test)])
```

```python
roc_auc_score(y_test, grid.predict_proba(x_test)[:,1], average='macro')
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, grid.predict(x_test)))
```

```python
from xgboost import plot_importance

fig, ax = plt.subplots(1,1, figsize=(10,8))
plot_importance(xgb_cls, ax=ax, max_num_features=20, height=0.4)
```
