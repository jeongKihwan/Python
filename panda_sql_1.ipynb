```python
# 대륙은 유럽이고, 인구수는 10,000,000 이상인 국가들을 추출
```

```python
import time
```

```python
start_time = time.time()
sql='select * from country where continent = "Europe" and population >=10000000;'

pd.read_sql(sql,db).head()

end_time = time.time()

execute_time = end_time - start_time
print(execute_time)
```

```python
# continent = 'europe' or population >= 10000000
sql='select * from country where continent = "Europe" or population >=10000000;'

pd.read_sql(sql,db).head()
```
