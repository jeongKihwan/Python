```python
# break, continue

# 1. break
#   반복문을 반복실행하다가 break가 실행되면 반복문을 빠져나오게 된다.

# 2. continue
#   반복문을 반복실행하다가 continue가 실행되면 continue 아래 코드를 실행하지 않고 
#   반복문의 처음으로 돌아와서 반복문을 계속 실행한다.
```

```python
# break를 사용해서 0~5까지만 출력..

i = 0
while i <= 10:
    print(i)
    i += 1
    if i > 5:
        break
```

```python
i = 0
while i < 10:
    i += 1
    if i%3 == 0:
        continue
    print(i)
```

```python
# 아래의 1~10까지 출력하는 코드에서 break와 continue를 활용해서 4,5,6,7만 출력하도록

i = 0

while i < 10:
    i += 1
    if i < 4:
        continue
    if i > 7:
        break
    print(i)
```

```python
# 글자의 길이가 4 이상인 단어만 리스트에 담으시오.

words = ['mountain', 'sea','sky','river','cloud']

L1 = []
for w in words:
    if len(w) > 4:
        L1.append(w)
L1
```

```python
# 글자의 길이가 4 이상인 단어의 수는 몇인가.

words = ['mountain', 'sea','sky','river','cloud']

count = 0
for w in words:
    if len(w) > 4:
        count += 1
count
```

```python
# words에서 알파벳순서에 따라 

# => "1번 : cloud"
# => "2번 : mountain"
# => "3번 : river"
# ...

for i, w in enumerate(sorted(words)):
    print(i+1,'번 :', w)    
```

```python
# 점수를 5회 입력받은 후
# => "점수 평균은 ***입니다"

suma = 0
count = 0

while count < 5:
    score = int(input('점수를 입력하세요'))
#     print(score, count)
    count += 1
    suma = suma + score
    
mean_score = suma/(count)
print("당신의 평균점수는 {}점입니다".format(round(mean_score)))
```

```python
# words를 
# => {'mountain':'산', 'sea':'바다', 'sky':'하늘', 'river':'강', 'cloud':'구름'}로
# 전환해서 출력

words = ['mountain', 'sea','sky','river','cloud']
kor = ['산','바다','하늘','강','구름']

nature = {}

for i in range(len(words)):
    nature[words[i]] = kor[i]
    
nature
```

```python
words = ['mountain', 'sea','sky','river','cloud']
kor = ['산','바다','하늘','강','구름']

nature = {}

for i,j in zip(words, kor):
    nature[i] = kor[j]
    
nature
```

```python
# 모음은 모음끼리, 모음이 아닌 문자열은 별도의 리스트에 각각 담아서 리턴하시오..

Text = "Python programming is useful"
vowel = 'aeiou'

vlist = []
clist = []

for i in Text:
    if i in vowel:
        vlist.append(i)
    else:
        clist.append(i)
        
print(vlist)
print(clist)
```
