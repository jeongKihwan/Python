```python
# 긴 문장을 'poem.txt' 파일에 쓴 후, 4가지 방법으로 읽어서 출력하시오..
```

```python
text = """
이 시의 화자는 이별의 슬픔을 감내하면서 임에 대한 애절한 사랑을 은근히 
드러내는 인종(忍從)적인 여성의 어조를 보이고 있다. 그러나 이러한 애상적인 
호소 속에 자신의 충격과 슬픔, 그리고 이별이 가져다줄 커다란 상처를 은근히 
드러냄으로써 임이 떠나지 않기를 바라는 간절한 심정을 전달하고 있는 것으로 
볼 수도 있다.
"""
```

```python
f = open('poem.txt', 'w')
f.write(text)
f.close()
```

```python
f = open('poem.txt', 'r')
lines = f.read()
print(lines)
f.close()
```

```python
f = open('poem.txt', 'r')
lines = f.readlines()
print(lines)
f.close()
```

```python
f = open('poem.txt', 'r')

line = f.readline()

while line:
    print(line)
    line = f.readline()
```

```python
# 메모장 만들기 
# 메모할 내용을 입력받은 후, 파일에 'memo.txt'이 누적해서 기록하고 동시에
# 전체 메모내용을 출력하는 코드 작성.

# 쓰기모드 'w' => 덮어쓰기 모드
# 쓰기모드 'a' => 이어쓰기
```

```python
text = input('메모할 내용을 입력하세요.')

f = open('memo.txt','a')
f.write(text)
f.close()

f = open('memo.txt','r')
lines = f.read()
f.close()

print()
print('='*20, "메모장 내용")
print(lines)
```
