```
from sklearn.model_selection import train_test_split
train_input, val_input, train_target, val_target= train_test_split(train_input, train_target, test_size=0.2, random_state=111)
```

```
# rnn이나 lstm, gru에 학습을 시킬 때, 입력값의 토큰의 갯수가 동일해야..
# 입력값의 길이를 강제로 통일시키는 함수.. : pad_sequences()
```

```
from tensorflow.keras.preprocessing.sequence import pad_sequences
train_seq = pad_sequences(train_input, maxlen=100)
```

```
val_seq = pad_sequences(val_input, maxlen=100)
```

```
# token => one_hot 변환
# ['python', 'is', 'fun']

#           'python'  'is'  'fun'
# 'python' =>  1,      0,     0    => [1,0,0]
# 'is' =>      0       1      0    => [0,1,0]
# 'fun' =>     0       0      1    => [0,0,1]
```

```
train_seq.shape
```

```
from tensorflow import keras
train_oh = keras.utils.to_categorical(train_seq)
```

```
len(train_oh[0][0])
```

```
train_oh.shape  # rnn, lstm, gru에 입력값으로 들어갈 shape => 3차원...
```

```
val_oh = keras.utils.to_categorical(val_seq)
```

```
from tensorflow import keras
model = keras.Sequential()

model.add(keras.layers.SimpleRNN(8, input_shape=(100, 500)))
model.add(keras.layers.Dense(1, activation='sigmoid'))
```

```
model.summary()
```

```
model.compile(optimizer='Adam', loss='binary_crossentropy', metrics = ['accuracy'])
history = model.fit(train_oh, train_target, epochs=20, batch_size=64, validation_data=(val_oh, val_target))
```
