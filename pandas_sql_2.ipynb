```python
# 'payment' table에서 월별 전체매출, 평균매출, 결제건수
```

```python
df = pd.read_sql('select * from payment', db)
df.head()
```

```python
df.info()
```

```python
df['month'] = df.payment_date.dt.month
```

```python
df.head()
```

```python
df.groupby('month')['amount'].sum()
```

```python
df.groupby('month')['amount'].mean()
```

```python
df.groupby('month')['customer_id'].count()
```

```python
sql = 'select month(payment_date) month, avg(amount) "평균매출", sum(amount) "총매출", count(payment_id) from payment group by month'
pd.read_sql(sql,db)
```

```python
groued = df.groupby('month')
```

```python
import numpy as np
groued.agg({'amount':[np.mean, np.sum], 'payment_id':len})
```
